### Acessando o Banco de Dados criado através de bibliotecas Python
Utilizaremos a função "create_engine" da biblioteca "sqlalchemy", que cria um objeto de conexão com um banco de dados relacional. Este objeto de conexão é usado para se comunicar com o banco de dados e executar consultas SQL. Ele segue o padrão  "SGBD://USER:SENHA@HOST/DATABASE" que deve ser preenchido com as informações do seu banco.

```python
from sqlalchemy import create_engine
import pandas as pd
SGBD = "postgresql+psycopg2"
USER = "postgres"
SENHA = "admin"
HOST = "localhost"
DATABASE = "projeto"

engine = create_engine(f"{SGBD}://{USER}:{SENHA}@{HOST}:5432/{DATABASE}")
engine.connect()
```

### Análise exploratória

Verificando quais categorias possuem mais produtos:

```python
query = """
SELECT 
    product_category_name AS Categoria,
    COUNT(product_category_name) AS quantidade
FROM
public.products
GROUP BY product_category_name
ORDER BY quantidade DESC
"""
pd.read_sql_query(query, con = engine)
```

A categoria que possui mais produtos é "cama_mesa_banho", com 3029 produtos.


---
Verificando ID do produto mais vendido:

```python
query = """
SELECT 
    product_id AS id_produto,
    count(product_id) AS numero_vendas

FROM
public.order_items
GROUP BY product_id
ORDER BY numero_vendas DESC
"""

pd.read_sql_query(query, con = engine)
```

O ID do produto mais vendido é "aca2eb7d00ea1a7b8ebd4e68314663af", com 527 vendas.

---
Verificando estado e cidade com mais clientes únicos:

```python
#Estado com mais clientes únicos
query = """
SELECT
    customer_state AS estado_cliente,
    COUNT(DISTINCT customer_unique_id) AS clientes
FROM
public.customers
GROUP BY customer_state
ORDER BY clientes DESC
"""

pd.read_sql_query(query, con=engine)
```

O estado com mais clientes únicos é SP, com 40302 clientes.

```python
#Cidade com mais clientes únicos
query = """
SELECT 
    customer_city AS cidade_cliente,
    customer_state AS estado_cliente,
    COUNT(DISTINCT customer_unique_id) AS clientes
FROM
public.customers
GROUP BY customer_city, customer_state
ORDER BY clientes DESC
"""
pd.read_sql_query(query, con = engine)
```

A cidade com mais clientes únicos é São Paulo, com 14984 clientes únicos.

---
Relacionando tabela "sellers" à tabela "order_items", para saber quais são e de onde são os vendedores que mais arrecadaram:

```python
query = """
SELECT
    s.seller_id,
    seller_state,
    seller_city,
    SUM(o.price) AS total_vendido
FROM
public.sellers AS s
LEFT JOIN public.order_items AS o
ON s.seller_id = o.seller_id
GROUP BY s.seller_id
ORDER BY total_vendido DESC
"""
pd.read_sql_query(query, con = engine)
```

O id do vendedor que mais arrecadou foi "4869f7a5dfa277a7dca6462dcf3b52b2	", da cidade de Guariba-SP, que vendeu um total de R$229.472,63.


---
Relacionando tabelas "products" e "order_items", para saber quais categorias de produtos que mais arrecadaram:

```python

query= '''
SELECT 
    p.product_category_name AS categoria_produto,
    SUM(o.price) AS valor_vendido

FROM public.products AS p
LEFT JOIN public.order_items AS o
ON p.product_id = o.product_id
GROUP BY p.product_category_name
ORDER BY valor_vendido DESC
'''
pd.read_sql_query(query, con = engine)
```

A categoria que mais arrecadou foi "beleza_saude", com o valor vendido de R$1.258.681,34.

---
Relacionando as tabelas "orders" e "customers", para saber a data do primeiro pedido de cada cliente:

```python

query='''
SELECT c.customer_unique_id AS id_cliente, 
	MIN(o.order_purchase_timestamp) AS data_primeiro_pedido
FROM public.orders AS o 
LEFT JOIN public.customers AS c 
ON c.customer_id = o.customer_id
GROUP BY c.customer_unique_id
ORDER BY c.customer_unique_id DESC
'''
pd.read_sql_query(query, con = engine)
```

---
Relacionando tabelas "customers", "orders" e "order_payments", para saber qual foi o cliente que mais gastou e de onde ele é:

```python

query ='''
SELECT
    c.customer_unique_id AS id_cliente,
    c.customer_city AS cidade,
    c.customer_state AS estado,
    SUM(op.payment_value) AS total_pago
FROM public.orders AS o
LEFT JOIN public.customers AS c
ON c.customer_id = o.customer_id
LEFT JOIN public.order_payments AS op
ON op.order_id = o.order_id
GROUP BY c.customer_unique_id, c.customer_city, c.customer_state
HAVING NOT SUM(op.payment_value) IS NULL
ORDER BY total_pago DESC
'''
pd.read_sql_query(query, con = engine)
```

Usando Subquery para descobrir valor médio pago por cliente:

```python
#Valor médio pago por clientes
query ='''
WITH gastos_clientes AS(
    SELECT
    c.customer_unique_id,
    c.customer_city,
    c.customer_state,
    SUM(op.payment_value) AS total_pago
FROM public.orders AS o
LEFT JOIN public.customers AS c
ON c.customer_id = o.customer_id
LEFT JOIN public.order_payments AS op
ON op.order_id = o.order_id
GROUP BY c.customer_unique_id, c.customer_city, c.customer_state
ORDER BY total_pago DESC
)
SELECT CAST(AVG(total_pago) AS NUMERIC(10, 2)) AS media_total_pago
FROM gastos_clientes

'''
pd.read_sql_query(query, con = engine)
```

O Valor médio pago por cliente foi de R$166,38.

---
Vendedores que possuem a nota média de suas reviews iguais a 5:

```python
query ='''
	SELECT 
    o.seller_id AS id_vendedor,
    AVG(review_score) AS nota_media
FROM public.reviews AS r
LEFT JOIN order_items AS o
ON o.order_id = r.order_id
GROUP BY o.seller_id
HAVING AVG(review_score) = 5
ORDER BY nota_media DESC

'''
pd.read_sql_query(query, con = engine)
```
